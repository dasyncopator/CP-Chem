```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = r"C:\Users\Tanay Mannikar\OneDrive - The University of Texas at Austin\CP Chem\0~1m, step 0.2m, 240KHz sampling\0.2m.csv"
signal = pd.read_csv(file, delimiter=",")
powerspectrum = np.abs(np.fft.fft(signal))**2
powercepstrum = np.abs(np.fft.ifft(np.log(powerspectrum)))
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8,3), sharex=True, dpi=200)
ax1.plot(powerspectrum, linewidth=0.5)
ax2.plot(cepstrum, linewidth=0.5)
plt.tight_layout()
plt.show()
np.shape(powerspectrum)
```
